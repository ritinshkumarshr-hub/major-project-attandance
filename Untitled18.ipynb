```python
df["overall_score"] = df[["math score", "reading score", "writing score"]].mean(axis=1)
```

```python
difficulty_index = 100 - df[["math score", "reading score", "writing score"]].mean()

difficulty_index
```

```python
df["consistency_score"] = df[["math score", "reading score", "writing score"]].std(axis=1)
```

```python
def risk_label(row):
    if row["overall_score"] < 40 or row["consistency_score"] > 15:
        return "High Risk"
    elif row["overall_score"] < 60:
        return "Medium Risk"
    else:
        return "Low Risk"

df["risk_category"] = df.apply(risk_label, axis=1)
```

```python
np.random.seed(42)
df["attendance"] = np.clip(
    50 + (df["overall_score"] * 0.5) + np.random.normal(0, 5, len(df)),
    60, 100
)
```

```python
attendance_corr = df["attendance"].corr(df["overall_score"])
attendance_corr
```
